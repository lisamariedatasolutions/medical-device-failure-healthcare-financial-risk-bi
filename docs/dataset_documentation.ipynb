# Dataset Documentation

## Project Name

Medical Device Failure & Healthcare Supply Chain Risk BI Project

## Dataset Strategy

This project uses two datasets. The primary dataset is a medical device failure dataset that supports analysis of device reliability, maintenance cost, downtime, manufacturer patterns, and failure event counts. The secondary dataset is a healthcare financial expense dataset that supports analysis of operating costs related to equipment, supplies, and staffing.

Together, the datasets support a business intelligence project focused on healthcare operations, medical device maintenance burden, downtime risk, and financial impact.

## Primary Dataset

The main dataset is:

Medical_Device_Failure_dataset.csv

This dataset contains medical device maintenance and failure-related records.

### Primary Dataset Fields

| Column | Description |
|---|---|
| Device_ID | Unique medical device identifier |
| Device_Type | Type of medical device |
| Purchase_Date | Date the device was purchased |
| Age | Device age |
| Manufacturer | Device manufacturer |
| Model | Device model |
| Country | Country associated with the device or manufacturer record |
| Maintenance_Cost | Cost associated with maintenance |
| Downtime | Device downtime value |
| Maintenance_Frequency | Frequency of maintenance |
| Failure_Event_Count | Number of failure events recorded |
| Maintenance_Class | Maintenance severity or classification level |
| Maintenance_Report | Text description of the maintenance or failure issue |

## Secondary Dataset

The financial support dataset is:

financial_data.csv

This dataset contains healthcare expense records by date, expense category, amount, and description.

### Secondary Dataset Fields

| Column | Description |
|---|---|
| Date | Expense transaction date |
| Expense_Category | Original category assigned to the expense |
| Amount | Expense amount |
| Description | Description of the expense item |

## Data Quality Notes

The medical device failure dataset has no missing values and no duplicate rows.

The financial dataset has no missing values and no duplicate rows. However, some expense descriptions may not match the assigned expense category.<br>
 For example, a row may list surgical masks under staffing or surgeon salaries under supplies. During cleaning, a corrected expense category field will be created based on the description.

## Planned Cleaned Files

The cleaned files will be exported as:

- medical_device_failure_cleaned.csv
- financial_data_cleaned.csv

## Planned Analysis Questions

This project will answer questions such as:

1. Which medical device types have the highest failure event counts?
2. Which device types have the highest average downtime?
3. Which manufacturers are associated with higher maintenance cost?
4. Which maintenance classes are most expensive?
5. Which older devices may need replacement review?
6. Which devices create the greatest operational risk?
7. Which healthcare expense categories account for the highest spending?
8. How do equipment, supply, and staffing expenses trend over time?
9. How can maintenance cost and downtime be connected to broader financial impact?

## Data Limitations

The datasets are used for portfolio-based business intelligence analysis. The project should not be interpreted as a clinical safety study,<br>
 a national medical device failure rate analysis, or an official hospital financial report.

The financial dataset is simplified and will require validation of category labels before analysis.

## Portfolio Purpose

This project demonstrates beginner-to-intermediate BI/data analyst skills in a healthcare operations context, including Python data cleaning, <br>
validation, exploratory analysis, SQL querying, Excel validation, Power BI dashboard design, PowerPoint executive reporting, and GitHub documentation.

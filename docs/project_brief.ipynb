# Medical Device Failure & Healthcare Supply Chain BI Project

## Business Problem

Hospitals and healthcare organizations rely on medical devices, replacement parts, suppliers, maintenance schedules, <br>
and inventory availability to support safe and efficient care delivery. When devices fail, replacement parts are delayed, <br> 
suppliers experience shortages, or repairs become too costly, healthcare operations can be disrupted. <br>

This project analyzes medical device failure and healthcare supply chain risk from a business intelligence perspective. <br>
The goal is to identify patterns in device failures, downtime, repair costs, supplier performance, and inventory risk so <br>
hospital operations teams can make better data-driven decisions. <br>

The project is designed as an end-to-end BI portfolio project using Python, SQL, Excel, Power BI, and PowerPoint.

## Financial Impact

Medical device failure and supply chain delays can increase healthcare operating costs through repair expenses, replacement part costs, vendor service fees, <br>
 emergency purchasing, and equipment replacement. When critical devices are unavailable, hospitals may also experience indirect financial consequences such as <br>
 delayed procedures, reduced department capacity, staff inefficiency, and increased reliance on rental or backup equipment.<br>

A BI dashboard can help leadership identify high-cost device categories, suppliers associated with delays, departments with repeated downtime, <br>
and devices that may require replacement instead of continued repair.

## Operational Impact

Medical device failures can disrupt hospital workflows when equipment is unavailable, repairs are delayed, or replacement parts are out of stock.<br>
 These disruptions may affect clinical departments, maintenance teams, procurement staff, and hospital leadership.<br>

This project supports operational decision-making by showing which device types fail most often, which departments experience the most downtime, <br>
which suppliers have longer delays, and where repair costs are becoming unusually high.

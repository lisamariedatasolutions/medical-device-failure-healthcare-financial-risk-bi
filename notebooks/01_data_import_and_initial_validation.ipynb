```python
# ==========================================
# Check missing values
# ==========================================


print("Missing values in device dataset:")
print(medical_device_df.isnull().sum())


print("\nMissing values in financial dataset:")    
print(financial_df.isnull().sum())

```

```python
# ==========================================
# Check duplicate rows
# ==========================================

print("Duplicate rows in device dataset:", medical_device_df.duplicated().sum())

print("Duplicate rows in financial dataset:", financial_df.duplicated().sum())

```

```python
# ==========================================
# Create raw dataset summary
# ==========================================

dataset_summary = pd.DataFrame({
    "Dataset": [
        "Medical Device Failure",
        "Financial Data"
    ],
    "File_Name": [
        "Medical_Device_Failure_dataset.csv",
        "financial_data.csv"
    ],
    "Row_Count": [
       medical_device_df.shape[0],
       financial_df.shape[0]
    ],
    "Column_Count": [
        medical_device_df.shape[1],
        financial_df.shape[1]
    ],
    "Duplicate_Rows": [
        medical_device_df.duplicated().sum(),
        financial_df.duplicated().sum()
    ],
    "Total_Missing_Values": [
        medical_device_df.isnull().sum().sum(),
        financial_df.isnull().sum().sum()
    ]
})

dataset_summary
```

```python
# ==========================================
# Export raw dataset summary
# ==========================================

dataset_summary.to_csv(r"C:\Users\Charl\OneDrive\Documents\GitHub\Medical Device Failure & Healthcare Supply Chain Risk BI Project\data\exports\dataset_summary.csv", index=False)

print("Raw dataset summary exported successfully.")
```

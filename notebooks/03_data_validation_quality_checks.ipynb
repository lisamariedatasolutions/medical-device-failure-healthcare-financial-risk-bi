## Notebook 03: Data Validation and Quality Checks

This notebook validates the cleaned medical device failure and financial datasets.<br>
It checks row counts, missing values, duplicates, date fields, numeric ranges, category values,<br>
 corrected financial categories, maintenance cost corrections, operational risk logic, and replacement review logic.

```python
# ==========================================
# Notebook 03: Data Validation and Quality Checks
# Medical Device Failure & Healthcare Financial Risk BI Project
# ==========================================


import pandas as pd 
import numpy as np
from pathlib import Path  


pd.set_option("display.max_columns", None)
pd.set_option("display.width", 120)   


PROJECT_ROOT = Path(r"C:\Users\Charl\OneDrive\Documents\GitHub\Medical Device Failure & Healthcare Supply Chain Risk BI Project")

raw_data_path = Path(
    r"C:\Users\Charl\OneDrive\Documents\GitHub\Medical Device Failure & Healthcare Supply Chain Risk BI Project\data\raw"
)

raw_files = [
    raw_data_path / "financial_data.csv",
    raw_data_path / "Medical_Device_Failure_dataset.csv"
]
raw_files 


cleaned_data_path =Path(r"C:\Users\Charl\OneDrive\Documents\GitHub\Medical Device Failure & Healthcare Supply Chain Risk BI Project\data\cleaned")

cleaned_files = [
    cleaned_data_path / "financial_data_cleaned.csv",
    cleaned_data_path / "medical_device_failure_cleaned.csv"
]
cleaned_files


exports_path =Path(r"C:\Users\Charl\OneDrive\Documents\GitHub\Medical Device Failure & Healthcare Supply Chain Risk BI Project\data\exports")

print("Libraries imported and paths set successfully")
```

```python
# ==========================================
# Load raw and cleaned datasets
# ==========================================

device_raw = pd.read_csv(raw_data_path/"Medical_Device_Failure_dataset.csv")
financial_raw = pd.read_csv( raw_data_path/"financial_data.csv")

device_clean = pd.read_csv(cleaned_data_path/"medical_device_failure_cleaned.csv")
financial_clean  =pd.read_csv(cleaned_data_path/"financial_data_cleaned.csv")

print("\nDevice raw shape:", device_raw.shape)
print("Device cleaned shape:", device_clean.shape)

print("\nFinancial raw shape:", financial_raw.shape)
print("Financial cleaned shape:", financial_clean.shape)
```

```python
# ==========================================
# Row count validation
# ==========================================

row_count_validation = pd.DataFrame({
    "Dataset": [
        "Medical Device Failure",
        "Financial Data"
    ],
    "Raw_Row_Count": [
        device_raw.shape[0],
        financial_raw.shape[0]
    ],
    "Cleaned_Row_Count": [
        device_clean.shape[0],
        financial_clean.shape[0]
    ]
})

row_count_validation["Rows_Match"] = np.where(
    row_count_validation["Raw_Row_Count"] == row_count_validation["Cleaned_Row_Count"],
    "Pass",
    "Review"
)

row_count_validation
```

```python
# ==========================================
# Missing value validation
# ==========================================

missing_validation = pd.DataFrame({
    "Dataset": [
        "Medical Device Failure",
        "Financial Data"
    ],
    "Total_Missing_Values": [
        device_clean.isnull().sum().sum(),
        financial_clean.isnull().sum().sum()
    ]
})

missing_validation["Missing_Value_Check"] = np.where(
    missing_validation["Total_Missing_Values"] == 0,
    "Pass",
    "Review"
)

missing_validation
```

```python
# ==========================================
# Duplicate row validation
# ==========================================

duplicate_validation = pd.DataFrame({
    "Dataset": [
        "Medical Device Failure",
        "Financial Data"
    ],

"Duplicate_Rows": [
    device_clean.duplicated().sum(),
    financial_clean.duplicated().sum()
]
})

duplicate_validation["Duplicate_Check"]  = np.where(
    duplicate_validation["Duplicate_Rows"] == 0,
    "Pass",
    "Review"
)

duplicate_validation
```

```python
# ==========================================
# Required column validation
# ==========================================

required_device_columns = [
    "device_id",
    "device_type",
    "purchase_date",
    "purchase_year",
    "age",
    "expected_age",
    "age_validation_flag",
    "manufacturer",
    "maintenance_cost",
    "maintenance_cost_original",
    "maintenance_cost_quality_flag",
    "downtime_hours",
    "maintenance_frequency",
    "failure_event_count",
    "failure_flag",
    "failure_status",
    "maintenance_class",
    "maintenance_severity",
    "failure_type",
    "risk_score",
    "operational_risk_level",
    "replacement_review_flag"
]

required_financial_columns = [
    "date",
    "expense_year",
    "expense_month",
    "expense_month_year",
    "expense_category",
    "corrected_expense_category",
    "expense_category_validation_flag",
    "amount",
    "description"
]

missing_device_columns = [col for col in required_device_columns if col not in device_clean.columns]
missing_financial_columns = [col for col in required_financial_columns if col not in financial_clean.columns]

print("Missing required device columns:", missing_device_columns)
print("Missing required financial columns:", missing_financial_columns)

column_validation = pd.DataFrame({
    "Dataset": ["Medical Device Failure", "Financial Data"],
    "Missing_Required_Columns": [
        len(missing_device_columns),
        len(missing_financial_columns)
    ],
    "Column_Check": [
        "Pass" if len(missing_device_columns) == 0 else "Review",
        "Pass" if len(missing_financial_columns) == 0 else "Review"
    ]
})

column_validation
```

```python
# ==========================================
# Date validation
# ==========================================

device_clean["purchase_date"] = pd.to_datetime(device_clean["purchase_date"], errors="coerce")
financial_clean["date"] = pd.to_datetime(financial_clean["date"], errors="coerce")

date_validation = pd.DataFrame({
    "Dataset": [
        "Medical Device Failure",
        "Financial Data"
    ],
    "Date_Column": [
        "purchase_date",
        "date"
    ],
    "Invalid_Date_Count": [
        device_clean["purchase_date"].isnull().sum(),
        financial_clean["date"].isnull().sum()
    ],
    "Minimum_Date": [
        device_clean["purchase_date"].min(),
        financial_clean["date"].min()
    ],
    "Maximum_Date": [
        device_clean["purchase_date"].max(),
        financial_clean["date"].max()
    ]
})

date_validation["Date_Check"] = np.where(
    date_validation["Invalid_Date_Count"] == 0,
    "Pass",
    "Review"
)

date_validation
```

```python
# ==========================================
# Numeric range validation
# ==========================================

numeric_range_validation = pd.DataFrame({
    "Check": [
        "Negative maintenance cost after cleaning",
        "Negative downtime hours",
        "Negative financial amount",
        "Negative failure event count",
        "Risk score outside 0-5"
    ],
    "Issue_Count": [
        (device_clean["maintenance_cost"] < 0).sum(),
        (device_clean["downtime_hours"] < 0).sum(),
        (financial_clean["amount"] < 0).sum(),
        (device_clean["failure_event_count"] < 0).sum(),
        (~device_clean["risk_score"].between(0, 5)).sum()
    ]
})

numeric_range_validation["Range_Check"] = np.where(
    numeric_range_validation["Issue_Count"] == 0,
    "Pass",
    "Review"
)

numeric_range_validation
```

```python
# ==========================================
# Maintenance cost correction validation
# ==========================================

negative_original_costs = (device_clean["maintenance_cost_original"] < 0).sum()
negative_cleaned_costs = (device_clean["maintenance_cost"] < 0).sum()
negative_cost_flags = (device_clean["maintenance_cost_quality_flag"] == "Negative Cost Corrected").sum()

maintenance_cost_validation = pd.DataFrame({
    "Metric": [
        "Original negative maintenance costs",
        "Negative maintenance costs after cleaning",
        "Negative cost correction flags"
    ],
    "Value": [
        negative_original_costs,
        negative_cleaned_costs,
        negative_cost_flags
    ]
})

maintenance_cost_validation
```

```python
# ==========================================
# Confirm negative cost flags match original negative values
# ==========================================

cost_flag_mismatches = device_clean[
    ((device_clean["maintenance_cost_original"] < 0) & 
     (device_clean["maintenance_cost_quality_flag"] != "Negative Cost Corrected"))
    |
    ((device_clean["maintenance_cost_original"] >= 0) & 
     (device_clean["maintenance_cost_quality_flag"] == "Negative Cost Corrected"))
]

print("Maintenance cost flag mismatches:", len(cost_flag_mismatches))
```

```python
# ==========================================
# Age validation
# ==========================================

age_validation = device_clean["age_validation_flag"].value_counts().reset_index()
age_validation.columns = ["Age_Validation_Flag", "Record_Count"]

age_validation
```

```python
# ==========================================
# Devices requiring age review
# ==========================================

age_review_records = device_clean[device_clean["age_validation_flag"] == "Review"]

print("Age records requiring review:", len(age_review_records))
```

```python
# ==========================================
# Device category validation
# ==========================================

device_category_validation = pd.DataFrame({
    "Category_Field": [
        "maintenance_severity",
        "failure_status",
        "operational_risk_level",
        "replacement_review_flag"
    ],
    "Unique_Values": [
        sorted(device_clean["maintenance_severity"].dropna().unique().tolist()),
        sorted(device_clean["failure_status"].dropna().unique().tolist()),
        sorted(device_clean["operational_risk_level"].dropna().unique().tolist()),
        sorted(device_clean["replacement_review_flag"].dropna().unique().tolist())
    ],
    "Unique_Count": [
        device_clean["maintenance_severity"].nunique(),
        device_clean["failure_status"].nunique(),
        device_clean["operational_risk_level"].nunique(),
        device_clean["replacement_review_flag"].nunique()
    ]
})

device_category_validation
```

```python
device_category_validation.to_csv(exports_path/"device_category_validation.csv", index=False)
print("Device category validation exported.")
```

```python
# ==========================================
# Failure type validation
# ==========================================

failure_type_validation = device_clean["failure_type"].value_counts().reset_index()
failure_type_validation.columns = ["Failure_Type", "Record_Count"]

failure_type_validation
```

```python
# ==========================================
# Financial category validation
# ==========================================

financial_category_validation = financial_clean.groupby(
    ["description", "corrected_expense_category"]
).size().reset_index(name="Record_Count")

financial_category_validation
```

```python
# ==========================================
# Financial correction count
# ==========================================

financial_correction_summary = financial_clean["expense_category_validation_flag"].value_counts().reset_index()
financial_correction_summary.columns = ["validation_flag", "Record_Count"]
financial_correction_summary
```

```python
# ==========================================
# Export financial category validation
# ==========================================

financial_category_validation.to_csv(exports_path / "financial_category_validation.csv", index=False)

print("Financial category validation exported.")
```

```python
# ==========================================
# Validate corrected financial category logic
# ==========================================

financial_logic_issues = financial_clean[
    (
        financial_clean["description"].str.contains("ventilator", case=False, na=False) &
        (financial_clean["corrected_expense_category"] != "Equipment")
    )
    |
    (
        financial_clean["description"].str.contains("mask", case=False, na=False) &
        (financial_clean["corrected_expense_category"] != "Supplies")
    )
    |
    (
        financial_clean["description"].str.contains("salary|salaries|surgeon", case=False, na=False) &
        (financial_clean["corrected_expense_category"] != "Staffing")
    )
]

print("Financial category logic issues:", len(financial_logic_issues))
```

```python
# ==========================================
# Risk level validation
# ==========================================

risk_logic_issues = device_clean[
    (
        (device_clean["risk_score"].between(0, 1)) &
        (device_clean["operational_risk_level"] != "Low")
    )
    |
    (
        (device_clean["risk_score"].between(2, 3)) &
        (device_clean["operational_risk_level"] != "Medium")
    )
    |
    (
        (device_clean["risk_score"].between(4, 5)) &
        (device_clean["operational_risk_level"] != "High")
    )
]

print("Risk level logic issues:", len(risk_logic_issues))
```

```python
# ==========================================
# Risk level summary
# ==========================================

risk_level_validation = device_clean.groupby(
    ["risk_score", "operational_risk_level"]
).size().reset_index(name="Record_Count")

risk_level_validation
```

```python
# ==========================================
# Export risk level validation
# ==========================================

risk_level_validation.to_csv(exports_path/"risk_level_validation.csv", index=False)
print("Risk level validation exported.")
```

```python
# ==========================================
# Replacement review validation
# ==========================================

replacement_validation = device_clean["replacement_review_flag"].value_counts().reset_index()
replacement_validation.columns = ["Replacement_Review_Flag", "Record_Count"]

replacement_validation
```

```python
# ==========================================
# Create final validation summary
# ==========================================

validation_summary = pd.DataFrame({
    "Validation_Check": [
        "Device row count preserved",
        "Financial row count preserved",
        "Device missing values",
        "Financial missing values",
        "Device duplicate rows",
        "Financial duplicate rows",
        "Required device columns present",
        "Required financial columns present",
        "Device dates valid",
        "Financial dates valid",
        "No negative maintenance costs after cleaning",
        "No negative downtime values",
        "No negative financial amounts",
        "Negative maintenance cost flags correct",
        "Device age validation passed",
        "Financial category correction logic passed",
        "Risk level logic passed",
        "Risk score range valid"
    ],
    "Issue_Count": [
        0 if device_raw.shape[0] == device_clean.shape[0] else 1,
        0 if financial_raw.shape[0] == financial_clean.shape[0] else 1,
        device_clean.isnull().sum().sum(),
        financial_clean.isnull().sum().sum(),
        device_clean.duplicated().sum(),
        financial_clean.duplicated().sum(),
        len(missing_device_columns),
        len(missing_financial_columns),
        device_clean["purchase_date"].isnull().sum(),
        financial_clean["date"].isnull().sum(),
        (device_clean["maintenance_cost"] < 0).sum(),
        (device_clean["downtime_hours"] < 0).sum(),
        (financial_clean["amount"] < 0).sum(),
        len(cost_flag_mismatches),
        len(age_review_records),
        len(financial_logic_issues),
        len(risk_logic_issues),
        (~device_clean["risk_score"].between(0, 5)).sum()
    ]
})

validation_summary["Status"] = np.where(
    validation_summary["Issue_Count"] == 0,
    "Pass",
    "Review"
)

validation_summary
```

```python
# ==========================================
# Export validation summary
# ==========================================

validation_summary.to_csv(exports_path/"validation_summary.csv", index=False)
print("Validation summary exported successfully.")
print(exports_path / "validation_summary.csv")
```

```python
# ==========================================
# Final validation status
# ==========================================

review_count = (validation_summary["Status"] == "Review").sum()

if review_count == 0:
    print("All major validation checks passed. Cleaned datasets are ready for EDA, SQL, Excel, and Power BI.")
else:
    print(f"{review_count} validation checks require review before moving forward.")
```

## Data Validation Summary

The cleaned medical device failure and financial datasets were validated before analysis.

Validation checks confirmed that row counts were preserved, duplicate rows were not introduced, required columns were present, date fields were valid, and no unexpected missing values were found.

The medical device dataset was also checked for negative maintenance costs, downtime ranges, age validation, maintenance severity categories, failure status categories, operational risk logic, and replacement review logic.

The financial dataset was checked for corrected expense categories. Expense descriptions were used to correct category labels for equipment, supplies, and staffing. The corrected category logic passed validation.

The cleaned and validated datasets are ready for exploratory data analysis, SQL querying, Excel validation, Power BI dashboarding, and executive reporting.
